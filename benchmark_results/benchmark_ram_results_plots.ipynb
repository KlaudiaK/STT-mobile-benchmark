## Zużycie pamięci RAM

```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

audio_files = [
    # Librispeech test clean
    {"filename": "1580-141083-0040", "length": 3.75, "category": "short", "dataset": "librispeech_clean"},
    {"filename": "6930-76324-0013", "length": 4.305, "category": "short", "dataset": "librispeech_clean"},
    {"filename": "8230-279154-0029", "length": 8.54, "category": "medium", "dataset": "librispeech_clean"},
    {"filename": "5683-32879-0011", "length": 9.21, "category": "medium", "dataset": "librispeech_clean"},
    {"filename": "2094-142345-0006", "length": 13.82, "category": "long", "dataset": "librispeech_clean"},
    {"filename": "7176-92135-0012", "length": 16.4, "category": "long", "dataset": "librispeech_clean"},

    # Librispeech test other
    {"filename": "4350-9170-0010", "length": 2.285, "category": "short", "dataset": "librispeech_other"},
    {"filename": "3764-168671-0043", "length": 3.8, "category": "short", "dataset": "librispeech_other"},
    {"filename": "1688-142285-0007", "length": 7.06, "category": "medium", "dataset": "librispeech_other"},
    {"filename": "6070-86744-0014", "length": 8.7, "category": "medium", "dataset": "librispeech_other"},
    {"filename": "367-293981-0018", "length": 12.3, "category": "long", "dataset": "librispeech_other"},
    {"filename": "7105-2330-0002", "length": 16.595, "category": "long", "dataset": "librispeech_other"},

    # Common Voice
    {"filename": "common_voice_en_276369", "length": 3.504, "category": "short", "dataset": "common_voice"},
    {"filename": "common_voice_en_39820939", "length": 3.42, "category": "short", "dataset": "common_voice"},
    {"filename": "common_voice_en_20025414", "length": 7.008, "category": "medium", "dataset": "common_voice"},
    {"filename": "common_voice_en_18272583", "length": 5.184, "category": "medium", "dataset": "common_voice"},
    {"filename": "common_voice_en_18725234", "length": 10.272, "category": "long", "dataset": "common_voice"},
    {"filename": "common_voice_en_19222168", "length": 10.296, "category": "long", "dataset": "common_voice"},
]

devices = {
    "note10": "Samsung Note9",
    "s23": "Samsung S23",
    "redmi": "Xiaomi Redmi"
}
technologies = ["sherpa_ncnn", "sherpa_onnx", "vosk"]

# Color mapping for categories
category_colors = {
    "short": "#2E8B57",
    "medium": "#FF8C00",
    "long": "#DC143C"
}

# Device colors for comparison plots
device_colors = {
    "note10": "#1f77b4",
    "s23": "#ff7f0e",
    "redmi": "#2ca02c"
}

def load_results(file_path):
    with open(file_path, "r") as f:
        return json.load(f)

def average_device_data_across_iterations(data, num_points=500):
    interpolated_rams = []
    min_time = float('inf')
    max_time = 0

    for iteration in data["iterations"]:
        times = [m["time"] for m in iteration["measures"]]
        min_time = min(min_time, min(times))
        max_time = max(max_time, max(times))

    common_times = np.linspace(min_time, max_time, num_points)

    for iteration in data["iterations"]:
        times = [m["time"] for m in iteration["measures"]]
        rams = [m["ram"] for m in iteration["measures"]]
        df = pd.DataFrame({"time": times, "ram": rams}).sort_values("time")
        interpolated = np.interp(common_times, df["time"], df["ram"])
        interpolated_rams.append(interpolated)

    avg_ram = np.mean(interpolated_rams, axis=0)
    return common_times, avg_ram

# 1. PLOTS BY DATASET
def create_dataset_plots():
    datasets = ["librispeech_clean", "librispeech_other", "common_voice"]
    dataset_names = {
        "librispeech_clean": "LibriSpeech Test Clean",
        "librispeech_other": "LibriSpeech Test Other",
        "common_voice": "Common Voice"
    }

    for tech in technologies:
        for device_short, device_full in devices.items():
            for dataset in datasets:
                plt.figure(figsize=(14, 8))
                files_found = False

                dataset_files = [audio for audio in audio_files if audio["dataset"] == dataset]

                for audio in dataset_files:
                    filename = audio["filename"]
                    category = audio["category"]
                    length = audio["length"]

                    result_filename = f"results_{device_short}_{filename}_{tech}.json"
                    print(f"Looking for: {result_filename}")

                    if os.path.exists(result_filename):
                        data = load_results(result_filename)
                        times, avg_ram = average_device_data_across_iterations(data)

                        label = f"{filename} ({category}, {length}s)"
                        color = category_colors[category]

                        plt.plot(times, avg_ram, label=label, color=color, linewidth=2, alpha=0.8)
                        files_found = True
                    else:
                        print(f"Missing file: {result_filename}")

                if files_found:
                    plt.title(f"Zużycie RAMu - {dataset_names[dataset]}\nModel: {device_full} | Technologia: {tech}", fontsize=16)
                    plt.xlabel("Czas (ms)", fontsize=12)
                    plt.ylabel("Zużycie RAMu (MB)", fontsize=12)
                    plt.legend(title="Nazwa pliku audio", fontsize=10, loc="best")
                    plt.ylim(bottom=0)
                    plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
                    plt.tight_layout()

                    output_filename = f"dataset_{dataset}_{tech}_{device_short}.png"
                    plt.savefig(output_filename, dpi=300, bbox_inches='tight')
                    print(f"Saved plot: {output_filename}")
                    plt.show()

                plt.close()

# 2. DEVICE COMPARISON PLOTS - Same technology on different devices
def create_device_comparison_plots():
    datasets = ["librispeech_clean", "librispeech_other", "common_voice"]
    dataset_names = {
        "librispeech_clean": "LibriSpeech Test Clean",
        "librispeech_other": "LibriSpeech Test Other",
        "common_voice": "Common Voice"
    }

    for tech in technologies:
        for dataset in datasets:
            # Filter audio files for current dataset
            dataset_files = [audio for audio in audio_files if audio["dataset"] == dataset]

            for audio in dataset_files:
                filename = audio["filename"]
                category = audio["category"]
                length = audio["length"]

                plt.figure(figsize=(14, 8))
                devices_found = False

                for device_short, device_full in devices.items():
                    result_filename = f"results_{device_short}_{filename}_{tech}.json"
                    print(f"Looking for: {result_filename}")

                    if os.path.exists(result_filename):
                        data = load_results(result_filename)
                        times, avg_ram = average_device_data_across_iterations(data)

                        label = f"{device_full}"
                        color = device_colors[device_short]

                        plt.plot(times, avg_ram, label=label, color=color, linewidth=2, alpha=0.8)
                        devices_found = True
                    else:
                        print(f"Missing file: {result_filename}")

                if devices_found:
                    plt.title(f"Zużycie RAMu - {filename} ({category}, {length}s)\nZbiór danych: {dataset_names[dataset]} | Technologia: {tech}", fontsize=16)
                    plt.xlabel("Czas (ms)", fontsize=12)
                    plt.ylabel("Zużycie RAMu (MB)", fontsize=12)
                    plt.legend(title="Model urządzenia", fontsize=12, loc="best")
                    plt.ylim(bottom=0)
                    plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
                    plt.tight_layout()

                    output_filename = f"device_comparison_{filename}_{tech}.png"
                    plt.savefig(output_filename, dpi=300, bbox_inches='tight')
                    print(f"Saved plot: {output_filename}")
                    plt.show()

                plt.close()

# 3. TECHNOLOGY COMPARISON PLOTS - Same device on different technologies
def create_technology_comparison_plots():
    datasets = ["librispeech_clean", "librispeech_other", "common_voice"]
    dataset_names = {
        "librispeech_clean": "LibriSpeech Test Clean",
        "librispeech_other": "LibriSpeech Test Other",
        "common_voice": "Common Voice"
    }

    # Technology colors
    tech_colors = {
        "sherpa_ncnn": "#e377c2",
        "sherpa_onnx": "#17becf",
        "vosk": "#bcbd22"
    }

    for device_short, device_full in devices.items():
        for dataset in datasets:
            dataset_files = [audio for audio in audio_files if audio["dataset"] == dataset]

            for audio in dataset_files:
                filename = audio["filename"]
                category = audio["category"]
                length = audio["length"]

                plt.figure(figsize=(14, 8))
                techs_found = False

                for tech in technologies:
                    result_filename = f"results_{device_short}_{filename}_{tech}.json"
                    print(f"Looking for: {result_filename}")

                    if os.path.exists(result_filename):
                        data = load_results(result_filename)
                        times, avg_ram = average_device_data_across_iterations(data)

                        label = f"{tech}"
                        color = tech_colors[tech]

                        plt.plot(times, avg_ram, label=label, color=color, linewidth=2, alpha=0.8)
                        techs_found = True
                    else:
                        print(f"Missing file: {result_filename}")

                if techs_found:
                    plt.title(f"Zużycie RAMu - {filename} ({category}, {length}s)\nModel urządzenia: {device_full} | Zbiór danych: {dataset_names[dataset]}", fontsize=16)
                    plt.xlabel("Czas (ms)", fontsize=12)
                    plt.ylabel("Zyżycie RAMu (MB)", fontsize=12)
                    plt.legend(title="Technology", fontsize=12, loc="best")
                    plt.ylim(bottom=0)
                    plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
                    plt.tight_layout()

                    output_filename = f"tech_comparison_{filename}_{device_short}.png"
                    plt.savefig(output_filename, dpi=300, bbox_inches='tight')
                    print(f"Saved plot: {output_filename}")
                    plt.show()

                plt.close()

if __name__ == "__main__":
    print("=== Generating Dataset Plots ===")
    create_dataset_plots()

    print("\n=== Generating Device Comparison Plots ===")
    create_device_comparison_plots()

    print("\n=== Generating Technology Comparison Plots ===")
    create_technology_comparison_plots()

    print("\nAll plots generated successfully!")
```
